```python
markets = ['Basketball_MatchBet','Basketball_Spread','Tabletennis_MatchBet','Tennis_MatchBet',
          'Snooker_MatchBet','Handball_MatchBet','Rugby_Handicap','Football_BTTS','Football_O/U_2.5',
          'Football_O/U_1.5', 'Football_O/U_0.5', 'Football_DrawNoBet', 'American_Football_MatchBet',
          'American_Football_Spread','Babminton_MatchBet','UFC_MatchBet',
          'Volleyball_MatchBet','GAA','Cricket_MatchBet','Boxing_DrawNoBet','Esports']
```

```python
test = []
```

```python
def page_scrape(market):
    
            x = datetime.datetime.now()
    
            #' + x.strftime("%Y-%m-%d") + '
    
            match_list = driver.find_elements_by_xpath('//tr[contains(@data-ng-class,"' + x.strftime("%Y-%m-%d") + '")]')
            #match_list = driver.find_elements_by_xpath('//tr[contains(@data-ng-class,"2020-04-06")]')
            
            if len(match_list) == 0:
                return
        
            match_list = ([value.text for value in match_list]) 
            match_list = [x.replace('All Odds', market) for x in match_list ]
            matches  = ([value.split("\n") for value in match_list ])
            
            for i in range(1,len(match_list)):
                if len(matches[i]) != 6:
                    return
                else:
                    matches[i].append(links_dict.get(market))
                    test.append(matches[i])
            
```

```python
for market in markets:
    driver.get(links_dict[market])
    sleep(2)
    page_scrape(market)
    sleep(2)
```

```python
driver.close()
```

```python
df = pd.DataFrame(test)
df.columns = ["Time", "Home", "Away", "Home Odds", "Away Odds","Market",'Link']

df = df[["Market", "Time", "Home", "Home Odds", "Away", "Away Odds",'Link']]
df
```

```python
df = df[(df['Time']!='IN PLAY')]
df['Home Odds'] = pd.eval(df['Home Odds']) + 1
df['Away Odds'] = pd.eval(df['Away Odds']) + 1
```

```python
home_favs_df = df[df['Home Odds']<df['Away Odds']]
home_favs_df = home_favs_df[((home_favs_df['Home Odds'] - 1) * home_favs_df['Away Odds']) > home_favs_df['Home Odds']]
home_favs_df['Home Unit Stake'] = 1
home_favs_df['Away Unit Stake'] = home_favs_df['Home Odds'] - 1
home_favs_df['Potential Profit'] = (home_favs_df['Away Unit Stake'] * home_favs_df['Away Odds']) - home_favs_df['Home Odds']
```

```python
away_favs_df = df[df['Home Odds']>df['Away Odds']]
away_favs_df = away_favs_df[(((away_favs_df['Away Odds'] - 1) * away_favs_df['Home Odds']) > away_favs_df['Away Odds'])]
away_favs_df['Home Unit Stake'] = away_favs_df['Away Odds'] - 1
away_favs_df['Away Unit Stake'] = 1
away_favs_df['Potential Profit'] = (away_favs_df['Home Unit Stake'] * away_favs_df['Home Odds']) - away_favs_df['Away Odds']
```

```python
df = pd.concat([home_favs_df, away_favs_df])
```

```python
risk_free = df[~((df['Home Odds']>2) & (df['Away Odds']>2))]
```

```python
gaurenteed = df[(df['Home Odds']>2) & (df['Away Odds']>2)]
gaurenteed['Home Unit Stake'] = 1
gaurenteed['Away Unit Stake'] = 1
gaurenteed['Potential Profit'] = 'Gaurenteed'
```

```python
df = pd.concat([risk_free, gaurenteed])
```
